## **Mtodo por desviacin estndar**

#### Variable tarjetas_credito

```python
fig=plt.figure(figsize = (5,3))
plt.boxplot(data1["tarjetas_credito"])
plt.title("Outliers de tarjetas_credito")
plt.show()
```

```python
y2 = data1["tarjetas_credito"]
Ls1 = y2.mean() + 2.5*y2.std()
Li1 = y2.mean() - 2.5*y2.std()
print("Limite superior permitido:", Ls1)
print("Limite inferior permitido:", Li1)
```

```python
#Encontramos outliers
outliers = data1[(y2>Ls1)|(y2<Li1)]
outliers
```

```python
#obtenemos los datos limpios
dataClean_tarjetas_credito = data1[(y2<=Ls1) & (y2>=Li1)]

#Diagrama de caja y bigote
dataClean_tarjetas_credito.plot.box(column = ["tarjetas_credito"], figsize = (5,3))
```

```python
dataClean_tarjetas_credito.to_csv("dataClean_tarjetas_credito.csv")
```

## **Mtodo quartilico**

#### Variable Ventas_precios_constantes

```python
#Diagrama de caja y bigote
#Rango intercuartilico
fig=plt.figure(figsize = (5,3))
plt.boxplot(data1["ventas_precios_corrientes"])
plt.title("Outliers de ventas_precios_corrientes")
plt.show()
```

```python
y = data1["ventas_precios_corrientes"]

percentile25 = y.quantile(0.25) #Q1
percentile75 = y.quantile(0.75) #Q3
iqr = percentile75 - percentile25

Limite_Superior_iqr = percentile75 + 1.5*iqr
Limite_Inferior_iqr = percentile25 - 1.5*iqr
print("Limire superior permitido", Limite_Superior_iqr)
print("Limire inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos outliers
outliers = data1[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers
```

```python
#obtenemos los datos limpios
dataClean_ventas_precios_corrientes = data1[(y<=Limite_Superior_iqr) & (y>=Limite_Inferior_iqr)]

#Diagrama de caja y bigote
fig=plt.figure(figsize = (5,3))
plt.boxplot(dataClean_ventas_precios_corrientes["ventas_precios_corrientes"])
plt.title("Outliers de ventas_precios_constantes")
plt.show()
```

```python
dataClean_ventas_precios_corrientes.to_csv("ventas_precios_corrientes.csv")
```

## Final

Se decide dejar el outlier "nuevo" presente en las variables anteriores al ser variables que representan ventas, y considerar que el valor atpico puede ser importante para el anlisis, en vez de un error de escritura, de acuerdo al tipo de variable.
Adems, sabemos que al eliminar los valores atpicos y al volver a aplicar una grfica de boxplot (que ocupa mtodo IQR) la medida se vuelve a calcular con los nuevos datos limpios, generando nuevos lmites y nuevos outliers, al volver a manipular los datos este proceso podra repetirse muchas veces al volverse a calcular. Por eso los dejamos.

La decisin entre qu mtodo utilizar se bas un poco en el tipop de outliers de cada variable, ya que las grficas de caja son con mtodo IQR y si aplicamos desviacin estndar el reultado no siempre encontrar los outliers deseados.

El mtodo intercuartlico se utiliz ms para aquellas variables que tenan outliers dispersos (extremos) +, muy sesgados o muy pegados al rango tpico. Ya que este mtodo funciona mejor para datos no normales ya que no se ve afectado por los datos atpicos, mientras que la desviacin estndar s se ve afectada por los valores de los datos atpicos y puede no notar la verdadera dispersin de los datos.


An as, se consider ambos mtodos y se eligi el mejor para la manipulacin de outliers.
