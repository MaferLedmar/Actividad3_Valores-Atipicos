## **Método por desviación estándar**

#### Variable tarjetas_credito

```python
fig=plt.figure(figsize = (5,3))
plt.boxplot(data1["tarjetas_credito"])
plt.title("Outliers de tarjetas_credito")
plt.show()
```

```python
y2 = data1["tarjetas_credito"]
Ls1 = y2.mean() + 2.5*y2.std()
Li1 = y2.mean() - 2.5*y2.std()
print("Limite superior permitido:", Ls1)
print("Limite inferior permitido:", Li1)
```

```python
#Encontramos outliers
outliers = data1[(y2>Ls1)|(y2<Li1)]
outliers
```

```python
#obtenemos los datos limpios
dataClean_tarjetas_credito = data1[(y2<=Ls1) & (y2>=Li1)]

#Diagrama de caja y bigote
dataClean_tarjetas_credito.plot.box(column = ["tarjetas_credito"], figsize = (5,3))
```

```python
dataClean_tarjetas_credito.to_csv("dataClean_tarjetas_credito.csv")
```

## **Método quartilico**

#### Variable Ventas_precios_constantes

```python
#Diagrama de caja y bigote
#Rango intercuartilico
fig=plt.figure(figsize = (5,3))
plt.boxplot(data1["ventas_precios_corrientes"])
plt.title("Outliers de ventas_precios_corrientes")
plt.show()
```

```python
y = data1["ventas_precios_corrientes"]

percentile25 = y.quantile(0.25) #Q1
percentile75 = y.quantile(0.75) #Q3
iqr = percentile75 - percentile25

Limite_Superior_iqr = percentile75 + 1.5*iqr
Limite_Inferior_iqr = percentile25 - 1.5*iqr
print("Limire superior permitido", Limite_Superior_iqr)
print("Limire inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos outliers
outliers = data1[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers
```

```python
#obtenemos los datos limpios
dataClean_ventas_precios_corrientes = data1[(y<=Limite_Superior_iqr) & (y>=Limite_Inferior_iqr)]

#Diagrama de caja y bigote
fig=plt.figure(figsize = (5,3))
plt.boxplot(dataClean_ventas_precios_corrientes["ventas_precios_corrientes"])
plt.title("Outliers de ventas_precios_constantes")
plt.show()
```

```python
dataClean_ventas_precios_corrientes.to_csv("ventas_precios_corrientes.csv")
```

## Final

Se decide dejar el outlier "nuevo" presente en las variables anteriores al ser variables que representan ventas, y considerar que el valor atípico puede ser importante para el análisis, en vez de un error de escritura, de acuerdo al tipo de variable.
Además, sabemos que al eliminar los valores atípicos y al volver a aplicar una gráfica de boxplot (que ocupa método IQR) la medida se vuelve a calcular con los nuevos datos limpios, generando nuevos límites y nuevos outliers, al volver a manipular los datos este proceso podría repetirse muchas veces al volverse a calcular. Por eso los dejamos.

La decisión entre qué método utilizar se basó un poco en el tipop de outliers de cada variable, ya que las gráficas de caja son con método IQR y si aplicamos desviación estándar el reultado no siempre encontrará los outliers deseados.

El método intercuartílico se utilizó más para aquellas variables que tenían outliers dispersos (extremos) +, muy sesgados o muy pegados al rango típico. Ya que este método funciona mejor para datos no normales ya que no se ve afectado por los datos atípicos, mientras que la desviación estándar sí se ve afectada por los valores de los datos atípicos y puede no notar la verdadera dispersión de los datos.
